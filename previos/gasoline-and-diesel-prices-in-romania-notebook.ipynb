# Load the data
Nothing fancy, just load the data, prin the size, head and tail. Also join the two DFs.

```python
# load the data
#gasDf = pd.read_csv("/kaggle/input/gasoline-and-diesel-prices-in-romania/gasolinePricesInRomania.csv")
gasDf = pd.read_csv("rum/gasolinePricesInRomania.csv")
#dieselDf = pd.read_csv("/kaggle/input/gasoline-and-diesel-prices-in-romania/dieselPricesInRomania.csv")
dieselDf = pd.read_csv("rum/dieselPricesInRomania.csv")

pd.set_option('display.max_columns', None)
pd.set_option('expand_frame_repr', False)

# do some prints of the counts
print("### Gasoline DF count")
print(gasDf.count())

print()
print("### Diesel DF count")
print(dieselDf.count())

# join the two data frames, not really important
fuelDf = pd.concat([gasDf, dieselDf["diesel price per liter in RON"]], axis=1).reindex(gasDf.index)

# print the head
print()
print("### head and tail data for the joined DF")
print(fuelDf.head())
print()
print(fuelDf.tail())
```

# Plot the data
Simple plot to show the pice evolution over time.

```python
# do a plot fo the fuel prices
plt.plot(fuelDf["date"], fuelDf["gasoline price per liter in RON"], color= "green", linewidth=3)
plt.plot(fuelDf["date"], fuelDf["diesel price per liter in RON"], color = "black", linewidth=3)
plt.legend(["Gasoline price", "Diesel price"])

dataSetSize = fuelDf["date"].size
plt.xticks(np.arange(0, dataSetSize, round((dataSetSize / 5)/10)*10))

plt.title("Average Gasoline and Diesel Prices in Romania", fontsize = 14)

plt.show()

print()
print("### ok, all done")
```
